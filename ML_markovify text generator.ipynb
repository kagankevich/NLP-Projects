```
import pandas as pd
import string
```

```
# reading lenta.ru news
news = pd.read_csv('raw_news.csv', sep=',')
```

```
# each news starts with date - crop that
clean_news = []
for new in news['text'] :
  clean_news.append(new[new.find('2017')+5:])

# use first 500n news only, to match the size of idiot corpus
crop_news = clean_news[:500]
```

```
# read dostoevsky's idiot
with io.open("idiot.txt", encoding='utf-8') as f:
    idiot = f.read().splitlines()
print('corpus length:', len(idiot))
```

```
news_df = pd.DataFrame({'text': clean_news, 'category': 'news'})
```

```
idiot_df = pd.DataFrame({'text': idiot, 'category': 'idiot'})
```

```
corpus_df = news_df.append(idiot_df, ignore_index = True)
```

```
corpus_df
```

```
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(corpus_df.text)
```

```
X_train, X_test, y_train, y_test = train_test_split(bow, corpus_df.category)
```

```
nb = MultinomialNB()
nb.fit(X_train, y_train)
```

```
from sklearn.metrics import classification_report
```

```
print(classification_report(y_test, nb.predict(X_test)))
```
